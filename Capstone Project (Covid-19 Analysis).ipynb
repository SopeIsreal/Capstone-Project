```python
#we need to extract Nigeria from the global daily stats table from John hopkins and combine the 3 data
#But first we need to make Country/Region the index
jh_confirmed.set_index('Country/Region', inplace = True)
jh_recovered.set_index('Country/Region', inplace = True)
jh_death.set_index('Country/Region', inplace = True) 
```

```python
#preview
jh_death.head(1)
```

```python
#Extracting Nigeria from the global dataset:
nig_global = []
for index, rows in jh_confirmed.iterrows():
    if index == 'Nigeria':
        nig_global.append(rows)
        
for index, rows in jh_recovered.iterrows():
    if index == 'Nigeria':
        nig_global.append(rows)
        
for index, rows in jh_death.iterrows():
    if index == 'Nigeria':
        nig_global.append(rows)

nig_global = pd.DataFrame(nig_global)
nig_global
```

```python
#renaming the rows to differentiate what Nigeria means:
ind = list(nig_global.index)
ind[0] = 'Nig_conf'
ind[1] = 'Nig_recov'
ind[2] = 'Nig_death'

nig_global.index = ind
```

```python
#preview
nig_global
```

```python
#drop irrelevant columns
nig_global = nig_global.drop(['Province/State', 'Lat', 'Long'], axis = 1)
```

```python
#transpose for better view
nig_global = nig_global.transpose()
```

```python
# Cleaning data on budget_data table 
#generating variance column and percentage variance column
#first lets view the table once more so we dont have to scroll up for reference
budget_data.head(3)
```

```python
budget_change = []
for index, row in budget_data.iterrows():
    diff = row[1] - row[2]
    budget_change.append(diff)
    
budget_data['Budget Change(Bn)'] = budget_change
budget_data.head(1)
```

```python
perc_change = []
for index, row in budget_data.iterrows():
    perc = (row[3]/row[1])*100
    perc2 = round(perc, 2)
    perc_change.append(perc2)
    
budget_data['Percentage Change'] = perc_change
budget_data.head(1)
```
